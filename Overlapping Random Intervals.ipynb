```python
import numpy as np
```

```python
s=100000 # number of experiments
success=0 # number of experiments where one of the intervals has overlaps with the rest
n=4 # number of intervals in each simulation

for z in range(0,s):
    a=np.zeros(shape=(n,2))
    for i in range(0,n):
        a[i]=np.random.rand(2)
    for i in range(0,n):
        contador=0
        for j in range(0,n): 
            if (a[i,0]<a[j,0] and a[j,0]<a[i,1]):
                contador+=1
            elif a[j,0]<a[i,0] and a[i,0]<a[j,1]:
                contador+=1
        if(contador==n-1):
            success+=1
prob=success/s
print(str(prob)+" probability that one of the intervals has overlaps with all the others")
```

Estimating probabilities for each number of intervals:

```python
k=20 # range of intervals
for n in range(2,k):
    s=100000 # number of experiments
    success=0 # number of experiments where one of the intervals has overlaps with the rest
    for z in range(0,s):
        
        a=np.zeros(shape=(n,2))
        for v in range(0,n):
            a[v]=np.random.rand(2)
            
        for i in range(0,n):
            contador=0
            for j in range(0,n): 
                if (a[i,0]<a[j,0] and a[j,0]<a[i,1]):
                    contador+=1
                elif (a[j,0]<a[i,0] and a[i,0]<a[j,1]):
                    contador+=1
                    
            if(contador==n-1):
                success+=1
                break
    prob=success/s
    print(str(prob)+" probability for "+str(n)+" intervals")
```
